```python
import pandas as pd
import json


dataset = []
with open('Digital_Music_5.json', 'r') as file:
    for line in file:
        try:
            data = json.loads(line)
            dataset.append(data)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")
```

```python
rating_list = []
for i in dataset:
    rating_list.append((i["overall"], i["reviewerID"], i["asin"]))
```
